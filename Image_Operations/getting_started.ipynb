```python
import cv2
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

Loading images


```python
image = cv2.imread("../img/аватарка.jpg")
```

Displaying


```python
def plt_image(title: str, image: str):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

plt_image("My photo", image)
```

Saving Images


```python
cv2.imwrite("../img/output.png", image)
```
