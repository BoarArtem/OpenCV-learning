```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from typing import Any
```

```python
def imshow(title: str, image: Any, size: int = 5):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w / h
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()


# image = cv2.imread("../img/аватарка.jpg")
```

 Making a black square

```python
image = np.zeros((512, 512, 3), np.uint8)
image_gray = np.zeros((512, 512), np.uint8)

imshow("Black Canvas - RGB", image)
imshow("Black Canvas - Grayscale", image)
```

```python
cv2.line(image, (0, 0), (100, 655), (255, 127, 0), 5)

imshow("Black Canvas with line", image)
```

```python
cv2.rectangle(image, (100, 100), (250, 2500), (255, 127, 0), 5)
imshow("Black Canvas with rectangle", image)
```
